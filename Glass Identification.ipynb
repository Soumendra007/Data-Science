# Project Description

The dataset describes the chemical properties of glass and involves classifying samples of glass using their chemical properties as one of six classes. The dataset was credited to Vina Spiehler in 1987. The study of classification of types of glass was motivated by criminological investigation. At the scene of the crime, the glass left can be used as evidence...if it is correctly identified!

The chemical compositions are measured as the weight percent in corresponding oxide.

**Attribute Information-**


1.	Id number: 1 to 214

2.	RI: refractive index

3.	Na: Sodium (unit measurement: weight percent in corresponding oxide, as are attributes 4-10)

4.	Mg: Magnesium

5.	Al: Aluminum

6.	Si: Silicon

7.	K: Potassium

8.	Ca: Calcium
    
9.	Ba: Barium

10.	Fe: Iron

11.	Type of glass: (class attribute)


•	1- building_windows_float_processed

•	2- building_windows_non_float_processed

•	3- vehicle_windows_float_processed

•	4- vehicle_windows_non_float_processed (none in this database)

•	5- containers

•	6- tableware

•	7- headlamps


There are 214 observations in the dataset. The dataset can be divided into window glass (classes 1-4) and non-window glass (classes 5-7).

# Removing Outliers

**Using zscore**

```python
out=df[["RI (Refractive Index)","Na (Sodium)","Al (Aluminum)","Si (Silicon)","K (Potassium)","Ca (Calcium)","Ba (Barium)","Fe (Iron)"]]
from scipy.stats import zscore
z=np.abs(zscore(out))
z
```

```python
#Locating outliers location
np.where(z>3)
```

```python
#Finding outliers location by giving their location
z.iloc[105,5]
```

# Creating outliers free Dataset

```python
new_df=df[(z<3).all(axis=1)]
new_df.shape
```

```python
print("Old DataFrame",df.shape[0])
print("New DataFrame",new_df.shape[0])
```

```python
print("Data lost parcentage:-",((df.shape[0]-new_df.shape[0])/df.shape[0])*100)
```

**So we lost 20 rows from our Dataset due to outliers.** Lets use IQR method to detect outliers and lets see much data we lost after getting a outliers free dataset using IQR method.

**Remove outliers using outliers**

```python
#Q3
Q3=out.quantile(.75)
```

```python
#Q1
Q1=out.quantile(.25)
```

```python
#IQR
IQR=Q3-Q1
new_df1=df[~((df<(Q1-1.5*IQR)) | (df>(Q3+1.5*IQR))).any(axis=1)]
```

```python
print("Old DataFrame",df.shape[0])
print("New DataFrame",new_df1.shape[0])
```

```python
print("Data lost parcentage:-",((df.shape[0]-new_df1.shape[0])/df.shape[0])*100)
```

**So you can clearly see in IQR method I lost too much data.So I will stick to my outliers free dataset that is new_df.**

# Reduceing skewness

```python
#Trying to reduce skewness using cube root method
new_df["RI (Refractive Index)"]=np.cbrt(new_df["RI (Refractive Index)"])
new_df["Mg (Magnesium)"]=np.cbrt(new_df["Mg (Magnesium)"])
new_df["Al (Aluminum)"]=np.cbrt(new_df["Al (Aluminum)"])
new_df["Ca (Calcium)"]=np.cbrt(new_df["Ca (Calcium)"])
new_df["Ba (Barium)"]=np.cbrt(new_df["Ba (Barium)"])
new_df["Fe (Iron)"]=np.cbrt(new_df["Fe (Iron)"])
```

```python
#Checking skewness again
new_df.skew()
```

**You can see RI (Refractive Index),Mg (Magnesium),Ca (Calcium),Ba (Barium),Fe (Iron) these columns are still not in acceptable range.**

```python
#Trying to reduce skewness using boxcox in Positive skewness column and using sqrt for negative skewness columns
from scipy.stats import boxcox
new_df["RI (Refractive Index)"]=boxcox(new_df["RI (Refractive Index)"],0)
new_df["Ca (Calcium)"]=boxcox(new_df["Ca (Calcium)"],0)
new_df["Ba (Barium)"]=boxcox(new_df["Ba (Barium)"],0)
new_df["Fe (Iron)"]=boxcox(new_df["Fe (Iron)"],0)

#For negative skewed column I am using sqrt
new_df["Mg (Magnesium)"]=np.sqrt(new_df["Mg (Magnesium)"])
```

```python
#Checking skewness again
new_df.skew()
```

**RI (Refractive Index) and Mg (Magnesium) are still out of acceptable range.Ba (Barium) and Fe (Iron) has NaN skewness.**

```python
#Trying to reducing skewness using log function +1 for these columns those are still out of acceptable range
new_df["RI (Refractive Index)"]=np.log1p(new_df["RI (Refractive Index)"])
new_df["Mg (Magnesium)"]=np.log1p(new_df["Mg (Magnesium)"])
new_df["Ba (Barium)"]=np.log1p(new_df["Ba (Barium)"])
new_df["Fe (Iron)"]=np.log1p(new_df["Fe (Iron)"])
```

```python
#Checking skewness again
new_df.skew()
```

**Still these columns are not in acceptable range,so I have to use Power Tranformar technique to these colums.**

```python
#using PowerTransformer to reduce the skewness
from sklearn.preprocessing import PowerTransformer
reshape_RI_column=new_df["RI (Refractive Index)"].values.reshape(-1,1)
reshape_Mg_column=new_df["Mg (Magnesium)"].values.reshape(-1,1)
reshape_Ba_column=new_df["Ba (Barium)"].values.reshape(-1,1)
reshape_Fe_column=new_df["Fe (Iron)"].values.reshape(-1,1)

pt=PowerTransformer("yeo-johnson")  #yeo-johnson method works on both positive and Negative method

#Now fitting the Transfomer
RI_column_transformed=pt.fit_transform(reshape_RI_column)
Mg_column_transformed=pt.fit_transform(reshape_Mg_column)
Ba_column_transformed=pt.fit_transform(reshape_Ba_column)
Fe_column_transformed=pt.fit_transform(reshape_Fe_column)


#Convert the transformed array back to the DataFrame
new_df["RI (Refractive Index)"]=pd.DataFrame(RI_column_transformed)
new_df["Mg (Magnesium)"]=pd.DataFrame(Mg_column_transformed)
new_df["Ba (Barium)"]=pd.DataFrame(Ba_column_transformed)
new_df["Fe (Iron)"]=pd.DataFrame(Fe_column_transformed)
```

```python
new_df.skew()
```

**Observation:-**

*1) Here you can see all column's are in acceptable range except Mg (Magnesium).*

*2) For Mg (Magnesium) we applied all the mothods to reduce this colums skewness but it can't be ,so can consider this column's skewness value is a threshold value.**
