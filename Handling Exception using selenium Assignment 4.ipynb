**1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:** 

**A) Rank** 

**B) Name** 

**C) Artist** 

**D) Upload date** 

**E) Views** 

**Answer:-**

```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
import requests
import pandas as pd
import time
```

```python
code=requests.get(" https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
#code

driver=webdriver.Chrome()
driver.get(" https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")

everything=[]

all_tag=driver.find_elements(By.XPATH,'//td')
for i in all_tag:
    all_text=i.text
    everything.append(all_text)
    if len(everything)==182:
        break

#Deleting two unnessary elements form the list
del everything[:2]

most_view=pd.DataFrame({"Rank":everything[0::6],"Video Name":everything[1::6],"Artist Name":everything[2::6],"Upload Date":everything[4::6],"Views":everything[3::6]})

most_view

```

```python
driver.close()
```

```python
#Load more info by clicking More Matches option

for i in range(3):
    try:
        
        more_matches=driver.find_element(By.XPATH,'//div[@class="col-lg-12 col-md-12 col-sm-12"]/button')
        more_matches.click()
        time.sleep(2)
    except NoSuchElementException:
        break
```

```python
Series=[]
Place=[]
Date=[]
Time=[]
```

```python
#Scraping Series names 
try:
    series_tag=driver.find_elements(By.XPATH,'//h5[@class="match-tournament-name ng-binding"]')
    for s in series_tag:
        s_text=s.text
        Series.append(s_text)
except NoSuchElementException:
    Series.append("--")

#Scraping place names     
try:
    place_tag=driver.find_elements(By.XPATH,'//div[@class="match-place ng-scope"]')
    for p in place_tag:
        p_text=p.text
        Place.append(p_text)
except NoSuchElementException:
    Place.append("--")    
    
#Scraping Date
try:
    date_tag=driver.find_elements(By.XPATH,'//div[@class="match-dates ng-binding"]')
    for d in date_tag:
        d_text=d.text
        Date.append(d_text)
except NoSuchElementException:
    Date.append("--")
    

#Scraping time
try:
    time_tag=driver.find_elements(By.XPATH,'//div[@class="match-time no-margin ng-binding"]')
    for t in time_tag:
        t_text=t.text
        Time.append(t_text)
except NoSuchElementException:
    Time.append("--")
```

```python
Fixture=pd.DataFrame({"Series Name":Series,"Place":Place,"Date":Date,"Time":Time})
Fixture
```

**6. Scrape the details of Highest selling novels.**

**A) Book name** 

**B) Author name** 

**C) Volumes sold** 

**D) Publisher** 

**E) Genre** 

Url - https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare

**Answer:-**

```python
driver=webdriver.Chrome()
driver.get("https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare")

info=[]

#Scraping whole table data
details=driver.find_elements(By.XPATH,'//td')
for i in details:
    d_text=i.text
    info.append(d_text)
    
info.pop(0)

Novels=pd.DataFrame({"Book Name":info[1::6],"Author Name":info[2::6],"Volumes Sold":info[3::6],"Publisher":info[4::6],"Genre":info[5::6]})
Novels
```
